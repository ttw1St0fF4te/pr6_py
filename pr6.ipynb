# Цель: визуализировать данные в виде графиков, используя датасет из прошлой практической. Продемонстрировать линейные графики, столбчатые, круговые, рассеяные диаграммы и гистограммы.

# Линейный график №1. В нем задается размер 38х10 при помощи figsize. По столбцам name и rating берутся первые 10 значений и задается название кривой через label = 'Рейтинг'. Само название графика задается через .title, название по X и по Y задаются через .xlabel и .ylabel. Сама отрисовка графика осуществляется благодаря .show()

```python
import matplotlib.pyplot as plt
import pandas as pd
anime = pd.read_csv('anime.csv')

plt.figure(figsize=(38, 10))
plt.plot(anime['name'].head(10), anime['rating'].head(10), label='Рейтинг')
plt.title('Статистика рейтинга аниме')
plt.xlabel('Название аниме')
plt.ylabel('Рейтинг аниме')
plt.legend()
plt.show()
```

# Линейный график №2. В нем также все размеры и данные берутся как и в линейном графике №1, за исключением того, что мы изменили окраску самой кривой при помощи color=, добавили в нее маркеры благодаря marker= и задали им размер с помощью marketsize=. А также создали новый график, который отмечает два самых высокооцениваемых тайтла желтым цветом

```python
plt.figure(figsize=(38, 10))
plt.plot(anime['name'].head(10), anime['rating'].head(10), color='black', marker='o', markersize=10, label='Рейтинг')
plt.plot(anime['name'].head(2), anime['rating'].head(2), color='yellow', marker='o', markersize=10, label='Рейтинг')
plt.title('Статистика рейтинга аниме')
plt.xlabel('Название аниме')
plt.ylabel('Рейтинг аниме')
plt.legend()
plt.show()
```

# Столбчатая диаграмма №1. Все тоже самое, что и в Линейном. Отличие заключается только в том, что при создании диаграммы используется .bar. В нем также задаются столбцы по X и по Y, из которых берутся первые 7 значений благодаря .head(), а также изменяются названия и цвет.

```python
plt.figure(figsize=(38, 10))

plt.bar(anime['name'].head(7), anime['episodes'].head(7), label='Кол-во эпизодов', color='grey')
plt.title('Статистика эпизодов аниме')
plt.xlabel('Название аниме')
plt.ylabel('Эпизоды аниме')
plt.legend()
plt.show()
```

# Столбчатая диаграмма №2. В отличие от первой столбчатой диаграммы мы выбираем данные из первых 20 значений из столбцов type и members также изменены названия и цвет

```python
plt.figure(figsize=(38, 10))

plt.bar(anime['type'].head(20), anime['members'].head(20), label='Кол-во зрителей', color='purple')
plt.title('Статистика зрителей')
plt.xlabel('Жанр')
plt.ylabel('Зрители')
plt.legend()
plt.show()
```

# Наложенный график (произвольный 3 Линейный и Столбчатый №1). Здесь поверх Столбчатой диаграммы №1 мы отрисовываем новый линейный график. Это происходит обыкновенным созданием двух графиков как в пунктах выше. В нем также задаются величины X и Y, сколько значений и откуда они берутся, названия, цвет и т.д.

```python
plt.figure(figsize=(38, 10))

plt.bar(anime['name'].head(7), anime['episodes'].head(7), label='Кол-во эпизодов', color='grey')
plt.plot(anime['name'].head(7), anime['episodes'].head(7), color='red', marker='o', markersize=10)
plt.title('Статистика эпизодов аниме')
plt.xlabel('Название аниме')
plt.ylabel('Эпизоды аниме')
plt.legend()
plt.show()
```

# Круговая диаграмма №1. На диаграмме отображается распределение первых 10 кол-тв зрителей на определенный тайтл. Сама диаграмма задается через .pie. Самое главное отличие, что названия беруться путем придания labels 10 первых значений из датасета из столбца name

```python
plt.figure(figsize=(38, 10))

plt.pie(anime['members'].head(10), labels=anime['name'].head(10))
plt.title('Распределение зрителей на аниме')
plt.show()
```

# В Круговой диаграмме №2 производим те же самые действия, что и в 1, только добавляем отображение процентов через autopct='%1.1f%%', которое показывает какую часть по отношению к другим тайтлам занимает кол-во зрителей 

```python
plt.figure(figsize=(38, 10))

plt.pie(anime['members'].head(10), labels=anime['name'].head(10), autopct='%1.1f%%')
plt.title('Распределение зрителей на аниме')
plt.show()
```

# Диаграммы рессеяния №1. Задается при помощи .scatter, где также нужно указать X и Y и цвет. В ней нам показывается зеленым точками к какому рейтингу относится кол-во зрителей

```python
plt.figure(figsize=(7, 5))

plt.scatter(anime['rating'].head(30), anime['members'].head(30), color='green')
plt.title('Связь между рейтингом и кол-вом зрителей у первых 30 тайтлов')
plt.xlabel('Рейтинг')
plt.ylabel('Зрители')
plt.show
```

# В Диаграмме рассеивания №2 изменилось отображение путем свапа X и Y местами. Теперь в диаграмме отображается связь между кол-вом зрителей и рейтингом

```python
plt.figure(figsize=(7, 5))

plt.scatter(anime['members'].head(30), anime['rating'].head(30), color='green')
plt.title('Связь между кол-вом зрителей и рейтингом у первых 30 тайтлов')
plt.xlabel('Зрители')
plt.ylabel('Рейтинг')
plt.show
```

# Гистограмма №1. Сами гистограммы позволяют оценить, как часто появляются определенные значения в наборе данных и как эти значения распределены. На этом примере можно увидеть как распределен какой-либо рейтинг. Это достигается путем создания самой гистограммы через .hist, где задается столбец rating, чье распределение будет выведено. Кол-во столбов задается через bins=значение. Также можно задать цвет самих столбов и их краев, и указать прозрачность цвета через alpha=

```python
plt.figure(figsize=(7, 5))

plt.hist(anime['rating'], bins=30, color='blue', edgecolor='purple', alpha=1, label='Распределение рейтинга аниме')
plt.legend()
plt.show()
```

# Гистограмма №2. На этом примере можно увидеть как распределены первые 20 кол-тв эпизодов. Различие от 1 гистограммы - смена цвета столбцов, названия, прозрачности и самого столбца, распределение которого совершается

```python
plt.figure(figsize=(10, 5))

plt.hist(anime['episodes'].head(20), bins=10, color='blue', edgecolor='purple', alpha=0.5, label='Распределение эпизодов аниме')
plt.legend()
plt.show()
```

# Гистограмма №3. На этом примере можно увидеть как распределены первые 30 кол-тв зрителей. Различие от 1 гистограммы - также смена цвета столбцов, названия, прозрачности и самого столбца, распределение которого совершается

```python
plt.figure(figsize=(10, 5))

plt.hist(anime['members'].head(30), bins=5, color='yellow', edgecolor='blue', alpha=0.7, label='Распределение зрителей аниме')
plt.legend()
plt.show()
```
